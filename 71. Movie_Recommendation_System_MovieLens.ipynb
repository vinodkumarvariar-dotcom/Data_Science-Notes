# Movie Recommendation System (Item-Based Collaborative Filtering)

This notebook demonstrates how to build a movie recommendation system using the MovieLens 100K dataset.

## 1️⃣ Dataset Structure

After extracting the MovieLens 100K dataset, you will find files like:

- **u.data** → User–Movie ratings
- **u.item** → Movie metadata
- **u.user** → User demographics

## 2️⃣ Import Required Libraries

```
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

**Explanation:**
- `pandas` and `numpy` are used for data manipulation.
- `cosine_similarity` is used to measure similarity between movies.

## 3️⃣ Load the Ratings Data (u.data)

```
ratings = pd.read_csv(
    'u.data',
    sep='\t',
    names=['user_id', 'movie_id', 'rating', 'timestamp']
)

ratings.head()
```

**Explanation:**
- Loads user ratings data.
- Each row represents a user rating for a movie.

## 4️⃣ Load Movie Information (u.item)

```
movies = pd.read_csv(
    'u.item',
    sep='|',
    encoding='latin-1',
    header=None
)

movies = movies[[0, 1]]
movies.columns = ['movie_id', 'title']
movies.head()
```

**Explanation:**
- Loads movie metadata.
- Keeps only movie ID and title for recommendations.

## 5️⃣ Merge Ratings with Movie Titles

```
data = pd.merge(ratings, movies, on='movie_id')
data.head()
```

**Explanation:**
- Merges ratings with movie titles.
- Creates a single dataset for analysis.

## 6️⃣ Create User–Item Matrix

```
user_movie_matrix = data.pivot_table(
    index='user_id',
    columns='title',
    values='rating'
)

user_movie_matrix.head()
```

**Explanation:**
- Rows represent users.
- Columns represent movies.
- Values are ratings given by users.

## 7️⃣ Handle Missing Values

```
user_movie_matrix_filled = user_movie_matrix.fillna(0)
```

**Explanation:**
- Missing ratings are replaced with 0.
- Required for similarity computation.

## 8️⃣ Item-Based Collaborative Filtering

```
movie_similarity = cosine_similarity(user_movie_matrix_filled.T)
```

**Explanation:**
- Computes cosine similarity between movies.
- Transpose is used to compare items instead of users.

## 9️⃣ Convert Similarity Matrix to DataFrame

```
movie_similarity_df = pd.DataFrame(
    movie_similarity,
    index=user_movie_matrix_filled.columns,
    columns=user_movie_matrix_filled.columns
)

movie_similarity_df.head()
```

**Explanation:**
- Converts similarity scores into a readable DataFrame.
- Rows and columns represent movie titles.

## 🔟 Movie Recommendation Function

```
def recommend_movies(movie_title, top_n=5):
    if movie_title not in movie_similarity_df:
        return 'Movie not found in dataset'
    
    similar_movies = (
        movie_similarity_df[movie_title]
        .sort_values(ascending=False)
        .iloc[1:top_n+1]
    )
    
    return similar_movies
```

**Explanation:**
- Finds movies most similar to a given movie.
- Excludes the movie itself from recommendations.

## 1️⃣1️⃣ Test the Recommendation System

```
recommend_movies('Star Wars (1977)', top_n=5)
```

**Explanation:**
- Recommends top 5 movies similar to *Star Wars (1977)*.
- Based on user rating patterns.
