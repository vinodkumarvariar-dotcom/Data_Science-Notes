# Students Performance Dataset - Descriptive Statistics

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```
num_df = df[['math score', 'reading score', 'writing score']]
num_df.head()
```

## Measures of Central Tendency

```
mean_vals = num_df.mean()
median_vals = num_df.median()
mode_vals = num_df.mode().iloc[0]
mean_vals, median_vals, mode_vals
```

## Measures of Dispersion

```
range_vals = num_df.max() - num_df.min()
variance_vals = num_df.var()
sd_vals = num_df.std()
cv_vals = (sd_vals / mean_vals) * 100
range_vals, variance_vals, sd_vals, cv_vals
```

## Skewness & Kurtosis

```
skew_vals = num_df.skew()
kurtosis_vals = num_df.kurt()
skew_vals, kurtosis_vals
```

## Visualizations

```
num_df.hist(figsize=(12,5), bins=20)
plt.show()
```

```
plt.figure(figsize=(10,6))
sns.boxplot(data=num_df)
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

## Summary Table

```
summary = pd.DataFrame({
    'Mean': mean_vals,
    'Median': median_vals,
    'Mode': mode_vals,
    'Range': range_vals,
    'Variance': variance_vals,
    'Std Dev': sd_vals,
    'CV (%)': cv_vals,
    'Skewness': skew_vals,
    'Kurtosis': kurtosis_vals
})
summary
```
