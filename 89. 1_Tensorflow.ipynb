#What is TensorFlow?
TensorFlow is an open-source machine learning (ML) and deep learning (DL) framework developed by Google. It is widely used for building, training, and deploying ML models, especially deep neural networks.


#What is tf.constant in TensorFlow?
In TensorFlow, tf.constant() is used to create immutable tensors, meaning their values cannot be changed after they are created.

#Uses of tf.constant in TensorFlow
1.Creating Fixed-Value Tensors



*   Used to store constant values that do not change during computation.





```python
import tensorflow as tf
a = tf.constant(5)
b = tf.constant(3)
c = a + b
print(c.numpy())  # Output: 8
```

2. Storing Model Parameters (Predefined Weights/Biases)



*   If certain values should remain unchanged during training, tf.constant can be used.
3. Defining Fixed Arrays or Matrices

Useful for setting up input data or masks in machine learning models.




```python
matrix = tf.constant([[1, 2], [3, 4]])
print(matrix.numpy())
```



4. Performance Optimization
*  Since tf.constant is immutable, TensorFlow optimizes it for faster execution.




5. Using in Graph Computation


*   TensorFlow uses computational graphs, and constants are often used as fixed nodes in graphs.

```python
import tensorflow as tf
matrix1=tf.constant([[1,2,3],[4,5,6]])
matrix2=tf.constant([[7,8],[9,10],[11,12]])
mat_mul=tf.matmul(matrix1,matrix2)

print("matrix:\n",mat_mul.numpy())
```

```python
transpose=tf.transpose(matrix1)
print("transpose:\n",transpose.numpy())
```

```python
#element wise operations
a=tf.constant([1,2,3])
b=tf.constant([4,5,6])
add=tf.add(a,b)
print("add:",add.numpy())
```

```python
mul=tf.multiply(a,b)
print("mul:",mul.numpy())
```

```python
#aggregation operations
tensor=tf.constant([1,2,3,4])
sum_val=tf.reduce_sum(tensor)
print("sum:",sum_val.numpy())
```

```python
mean_value=tf.reduce_mean(tensor)
print("mean:",mean_value.numpy())
```

```python
tensor = tf.constant([1, 2, 3, 4], dtype=tf.float32)

mean_value = tf.reduce_mean(tensor)
print("mean:", mean_value.numpy())
```

```python
max_value=tf.reduce_max(tensor)
print("max:",max_value.numpy())
```

```python
import tensorflow as tf

# Define a constant tensor (e.g., input data)
input_data = tf.constant([1.0, 2.0, 3.0, 4.0])

# Define a variable (e.g., a weight that we can update during training)
weight = tf.Variable(0.5)

# Example computation: multiply input by weight
output = input_data * weight

print("Input Data:", input_data.numpy())
print("Weight:", weight.numpy())
print("Output:", output.numpy())

# Now we can update the weight
# For example, during a training step, we adjust the weight
weight.assign_add(0.1)  # Increase weight by 0.1

print("Updated Weight:", weight.numpy())
```
