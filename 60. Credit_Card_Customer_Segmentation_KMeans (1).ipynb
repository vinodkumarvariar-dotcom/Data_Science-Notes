# Credit Card Customer Segmentation using k-Means

## Dataset
Kaggle: Credit Card Dataset for Clustering (CC_GENERAL.csv)

Each row represents one customer and each column represents credit behavior.

## Business Objective
Segment customers into Premium, Risky, Low Engagement, and Average groups to enable targeted marketing, credit risk control, and personalized offers.

```
import pandas as pd

df = pd.read_csv('CC_GENERAL.csv')
df.head()
```

```
df.info()
df.describe()
```

### Key Observations
- Missing values present
- Features have very different scales
- Scaling is mandatory for k-Means

## Handle Missing Values
Median is used as it is robust to outliers.

```
df.isnull().sum()
```

```
df.fillna(df.median(), inplace=True)
```

## Feature Selection
We select business-relevant features to improve interpretability.

```
features = df[['CREDIT_LIMIT', 'PURCHASES_FREQUENCY', 'CASH_ADVANCE', 'PAYMENTS']]
```

## Feature Scaling
Scaling is critical because k-Means is distance-based.

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

## Elbow Method to Find Optimal k

```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_features)
    inertia.append(km.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

Elbow observed around **k = 4**

## Apply k-Means Clustering

```
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

df['Cluster'] = clusters
```

## Cluster Analysis

```
cluster_summary = df.groupby('Cluster')[
    ['CREDIT_LIMIT', 'PURCHASES_FREQUENCY', 'CASH_ADVANCE', 'PAYMENTS']
].mean()

cluster_summary
```

## Business Interpretation
- **Cluster 0 – Premium Customers**: High limit, high payments, low risk
- **Cluster 1 – Risky Customers**: High cash advance, low payments
- **Cluster 2 – Low Engagement**: Low usage and spending
- **Cluster 3 – Average Customers**: Balanced behavior

## PCA Visualization (Optional)

```
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_features)

plt.scatter(pca_data[:,0], pca_data[:,1], c=df['Cluster'])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Customer Segments Visualization')
plt.show()
```

## Final Outcome
- Successfully segmented customers into 4 meaningful groups
- Identified premium vs risky customers
- Delivered actionable financial insights
