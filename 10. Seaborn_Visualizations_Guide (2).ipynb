# Seaborn Visualizations Guide

This notebook demonstrates various plots available in the Seaborn library. 
We will use built-in datasets like `tips`, `iris`, `flights`, and `titanic`.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set the theme for better aesthetics
sns.set_theme(style="darkgrid")

# Load datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
flights = sns.load_dataset("flights")
```

**Code Explanation:**

*   **Import Libraries**: Imports `seaborn`, `matplotlib.pyplot`, `numpy`, and `pandas` for data visualization and manipulation.
*   **Set Theme**: Sets the Seaborn theme to `darkgrid` for better aesthetics.
*   **Load Datasets**: Loads four built-in datasets: `tips`, `iris`, `titanic`, and `flights`.

## 1. Count Plot

A **Count Plot** can be thought of as a histogram across a categorical, instead of quantitative, variable. The basic API and options are identical to those for `barplot()`, so you can compare counts across nested variables.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="class", data=titanic, palette="viridis")
plt.title("Count of Passengers by Class")
plt.show()
```

**Code Explanation:**

*   **Dataset**: Uses the `titanic` dataset.
*   **Plot Type**: Creates a count plot to visualize the frequency of passengers in each class.
*   **X-Axis**: Represents the `class` variable (First, Second, Third).
*   **Palette**: Applies the `viridis` color map for styling.

## 2. Bar Plot

A **Bar Plot** represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="day", y="total_bill", data=tips, palette="magma")
plt.title("Average Total Bill by Day")
plt.show()
```

**Code Explanation:**

*   **Dataset**: Uses the `tips` dataset.
*   **Plot Type**: Generates a bar plot showing the average `total_bill`.
*   **Axes**: `day` is on the x-axis, and `total_bill` is on the y-axis.
*   **Error Bars**: The vertical lines indicate the confidence interval around the mean.

## 3. Scatter Plot

A **Scatter Plot** depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. It allows you to see patterns and relationships.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips, style="time", s=100)
plt.title("Total Bill vs Tip")
plt.show()
```

**Code Explanation:**

*   **Variables**: Plots `total_bill` (x-axis) vs. `tip` (y-axis).
*   **Grouping**: Uses `hue='time'` to color points by time of day (Lunch/Dinner).
*   **Styling**: Uses `style='time'` to differentiate markers by time.
*   **Size**: Sets the point size to 100 with `s=100`.

## 4. Line Plot

A **Line Plot** is used to visualize the trend of a variable over time or another continuous variable. It is great for time series data.

```python
plt.figure(figsize=(10, 5))
# Aggregating data to show trend over years
sns.lineplot(x="year", y="passengers", data=flights, marker="o")
plt.title("Number of Passengers Over Years")
plt.show()
```

**Code Explanation:**

*   **Dataset**: Uses the `flights` dataset.
*   **Trend**: Visualizes the trend of `passengers` over `years`.
*   **Aggregation**: Automatically aggregates data (mean) for each year.
*   **Confidence Interval**: The shaded area represents the confidence interval.
*   **Markers**: Adds circular markers (`o`) to data points.

## 5. Box Plot

A **Box Plot** (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. It shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="coolwarm")
plt.title("Distribution of Total Bill by Day and Smoker Status")
plt.show()
```

**Code Explanation:**

*   **Distribution**: Shows the distribution of `total_bill` by `day`.
*   **Comparison**: Uses `hue='smoker'` to compare smokers vs. non-smokers.
*   **Components**: Displays the median (line), IQR (box), and range (whiskers).

## 6. Histogram

A **Histogram** represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=tips, x="total_bill", kde=True, bins=20, color="teal")
plt.title("Histogram of Total Bill")
plt.show()
```

**Code Explanation:**

*   **Frequency**: Shows the frequency distribution of `total_bill`.
*   **KDE Overlay**: `kde=True` adds a smooth density curve.
*   **Bins**: Divides the data into 20 bins (`bins=20`).

## 7. Density Plot (KDE Plot)

A **Density Plot** (Kernel Density Estimate) is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. It represents the data using a continuous probability density curve.

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=iris, x="sepal_width", hue="species", fill=True, alpha=0.5)
plt.title("Density Plot of Sepal Width by Species")
plt.show()
```

**Code Explanation:**

*   **Variable**: Visualizes the density of `sepal_width`.
*   **Grouping**: Separates distributions by `species`.
*   **Appearance**: Fills the area under the curve (`fill=True`) and adds transparency (`alpha=0.5`).

## 8. Violin Plot

A **Violin Plot** plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="day", y="total_bill", data=tips, palette="muted", split=True)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

**Code Explanation:**

*   **Hybrid Plot**: Combines box plot and KDE plot features.
*   **Data**: Shows `total_bill` distribution for each `day`.
*   **Shape**: The width of the violin represents data density at that value.

## 9. Swarm Plot

A **Swarm Plot** is similar to a strip plot, but the points are adjusted (only along the categorical axis) so that they don't overlap. This gives a better representation of the distribution of values, although it does not scale well to large numbers of observations.

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(x="day", y="total_bill", data=tips, color="black", alpha=0.7)
# Often combined with a boxplot or violinplot
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("Swarm Plot with Box Plot Overlay")
plt.show()
```

**Code Explanation:**

*   **Combination**: Overlays a `swarmplot` on a `boxplot`.
*   **Swarm Plot**: Shows individual data points without overlap.
*   **Box Plot**: Provides summary statistics (median, quartiles).
*   **Insight**: Offers both detailed data view and statistical summary.

## 10. Heatmap

A **Heatmap** is a way of representing the data in a 2-dimensional form. The data values are represented as colors in the graph. The goal of the heatmap is to provide a colored visual summary of information.

```python
plt.figure(figsize=(8, 6))
# Calculate the correlation matrix
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

**Code Explanation:**

*   **Correlation Matrix**: Calculates the correlation coefficients between all numeric variables in the `iris` dataset using `iris.corr(numeric_only=True)`.
*   **Visualization**: Visualizes the correlation matrix as a heatmap.
*   **Color Map**: `cmap='coolwarm'` diverges from cool (negative correlation) to warm (positive correlation).
*   **Annotations**: `annot=True` displays the correlation values on the heatmap.

## 11. Pair Plot

A **Pair Plot** plots pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column.

```python
sns.pairplot(iris, hue="species", palette="husl")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

**Code Explanation:**

*   **Grid**: Creates a matrix of scatter plots for all numeric variable pairs.
*   **Diagonal**: Shows univariate distribution (KDE/Histogram) for each variable.
*   **Clustering**: `hue='species'` colors points to show species separation.
