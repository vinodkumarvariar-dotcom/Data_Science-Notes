1️⃣ ONE-SAMPLE Z-TEST (MEAN)

📌 Problem

A company claims the average battery life is 10 hours.
A sample of 40 batteries has a mean life of 9.6 hours.
Population standard deviation is assumed known.
Test at 5% significance level.

🔹 Hypotheses

H₀: μ = 10

H₁: μ ≠ 10

| Part                | Meaning                    |
| ------------------- | -------------------------- |
| `statsmodels`       | Library                    |
| `statsmodels.stats` | Sub-package for statistics |
| `weightstats`       | **Module** (Python file)   |
| `ztest`             | Function inside the module |


```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data
data = np.array([
    9.5, 9.6, 9.4, 9.7, 9.6, 9.5, 9.7, 9.6, 9.4, 9.6,
    9.5, 9.6, 9.4, 9.7, 9.6, 9.5, 9.7, 9.6, 9.4, 9.6,
    9.5, 9.6, 9.4, 9.7, 9.6, 9.5, 9.7, 9.6, 9.4, 9.6
])

alpha = 0.05

z_stat, p_value = ztest(data, value=10)

print("One-Sample Z-Test")
print("Z-statistic:", z_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject H₀ → Mean is significantly different from 10")
else:
    print("Fail to Reject H₀ → No significant difference from 10")
```

2️⃣ TWO-SAMPLE Z-TEST (MEANS)
📌 Problem

Compare average marks of Class A and Class B.


Hypotheses (Two-Sample Z-Test for Means)

H₀ (Null Hypothesis):
μ₁ = μ₂
The mean score of Class A is equal to the mean score of Class B (no significant difference).

H₁ (Alternative Hypothesis):
μ₁ ≠ μ₂
The mean scores of Class A and Class B are different (there is a significant difference).

Significance Level: α = 0.05

```python
from statsmodels.stats.weightstats import ztest

class_A = [78, 85, 88, 90, 84, 86, 87, 89]
class_B = [72, 75, 70, 68, 74, 73, 71, 69]

alpha = 0.05

z_stat, p_value = ztest(class_A, class_B)

print("\nTwo-Sample Z-Test")
print("Z-statistic:", z_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject H₀ → The two means are significantly different")
else:
    print("Fail to Reject H₀ → No significant difference between means")
```

Z-TEST FOR PROPORTIONS (A/B TESTING)
📌 Problem

Compare conversion rates of Website A and Website B.

Hypotheses (Two-Sample Proportion Z-Test)

H₀ (Null Hypothesis):
p₁ = p₂
The conversion rates of Version A and Version B are equal (no significant difference).

H₁ (Alternative Hypothesis):
p₁ ≠ p₂
The conversion rates of Version A and Version B are different (there is a significant difference).

Significance Level: α = 0.05

```python
from statsmodels.stats.proportion import proportions_ztest

# Number of conversions
success = [200, 260]

# Total visitors
nobs = [1000, 1000]

alpha = 0.05

z_stat, p_value = proportions_ztest(success, nobs)

print("\nProportion Z-Test")
print("Z-statistic:", z_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject H₀ → Conversion rates are significantly different")
else:
    print("Fail to Reject H₀ → No significant difference in conversion rates")
```

ONE-SAMPLE T-TEST
📌 Problem

A university claims the average exam score is 70.
A sample of students has the following scores.
Test at 5% significance level whether the average score differs from 70.

🔹 Hypotheses

H₀ (Null): μ = 70

H₁ (Alternative): μ ≠ 70

```python
from scipy.stats import ttest_1samp

scores = [68, 72, 75, 70, 69, 71, 74]
alpha = 0.05

t_stat, p_value = ttest_1samp(scores, popmean=70)

print("One-Sample T-Test")
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject H₀")
    print("Conclusion: The average score is significantly different from 70.")
else:
    print("Fail to Reject H₀")
    print("Conclusion: No significant difference from 70.")
```

INDEPENDENT (TWO-SAMPLE) T-TEST
📌 Problem

Compare the average exam scores of Class A and Class B.
Test if the two class means are different.

🔹 Hypotheses

H₀: μ₁ = μ₂

H₁: μ₁ ≠ μ₂

```python
from scipy.stats import ttest_ind

class_A = [78, 85, 88, 90, 84]
class_B = [72, 75, 70, 68, 74]

alpha = 0.05

t_stat, p_value = ttest_ind(class_A, class_B)

print("\nIndependent Two-Sample T-Test")
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject H₀")
    print("Conclusion: The two class means are significantly different.")
else:
    print("Fail to Reject H₀")
    print("Conclusion: No significant difference between class means.")
```

PAIRED T-TEST
📌 Problem

A fitness program measures weights before and after a diet plan.
Test whether the diet caused a significant change in weight.



Correct Hypotheses (Paired T-Test)

Let


d=before−after

Null Hypothesis (H₀)
​

H0​:μd​=0

➡️ The mean weight before and after the diet is the same
➡️ The diet has no effect

Alternative Hypothesis (H₁)
H1​:μd​!=0

➡️ There is a significant change in weight due to the diet
➡️ (Two-tailed test)

```python
from scipy.stats import ttest_rel

# Weight data (paired observations)
before = [80, 82, 85, 83, 81]
after  = [78, 80, 82, 81, 79]

alpha = 0.05

# Paired t-test
t_stat, p_value = ttest_rel(before, after)

print("Paired T-Test Results")
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision rule
if p_value < alpha:
    print("Reject H₀")
    print("Conclusion: The diet caused a statistically significant change in weight.")
else:
    print("Fail to Reject H₀")
    print("Conclusion: There is no statistically significant change in weight due to the diet.")
```

Chi-Square Test of Independence

📌 Problem

A company wants to know whether gender is related to product preference.




|        | Product A | Product B |
| ------ | --------- | --------- |
| Male   | 30        | 20        |
| Female | 15        | 35        |


Hypotheses

H₀ (Null Hypothesis): Gender and product preference are independent  
H₁ (Alternative Hypothesis): Gender and product preference are associated (dependent)

```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table:
# Rows: Gender (Male, Female)
# Columns: Product Preference (A, B)
data = np.array([
    [30, 20],  # Male
    [15, 35]   # Female
])

alpha = 0.05

# Perform Chi-Square Test of Independence
chi2, p_value, dof, expected = chi2_contingency(data)

print("Chi-Square Test of Independence")
print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p_value)


# Decision and Conclusion
if p_value < alpha:
    print("\nDecision: Reject H₀")
    print("Conclusion: Gender and product preference are associated.")
else:
    print("\nDecision: Fail to Reject H₀")
    print("Conclusion: Gender and product preference are independent.")
```

Chi-Square Goodness-of-Fit Test

📌 Problem

A dice manufacturer claims the dice is fair.
Observed outcomes from 60 rolls:

Face	Observed
1–6	[8, 9, 10, 11, 12, 10]

Expected frequency = 10 each

🔹 Hypotheses

H₀: Observed distribution follows expected distribution

H₁: Observed distribution does not follow expected distribution

```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([8, 9, 10, 11, 12, 10])
expected = np.array([10, 10, 10, 10, 10, 10])

alpha = 0.05

chi2, p_value = chisquare(f_obs=observed, f_exp=expected)

print("\nChi-Square Goodness-of-Fit Test")
print("Chi-square statistic:", chi2)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject H₀")
    print("Conclusion: The dice is NOT fair.")
else:
    print("Fail to Reject H₀")
    print("Conclusion: The dice appears to be fair.")
```

F-TEST USING scipy.stats.f

🔹 Problem

Two machines A and B produce rods.
Test whether their variances are equal at α = 0.05.

🔹 Hypotheses

H₀: σ₁² = σ₂² (Equal variances)

H₁: σ₁² ≠ σ₂² (Unequal variances)

```python
import numpy as np
from scipy.stats import f

# Sample data
machine_A = [12, 14, 13, 15, 14, 16]
machine_B = [10, 11, 9, 8, 10, 9]

alpha = 0.05

# Sample variances (unbiased)
var_A = np.var(machine_A, ddof=1)
var_B = np.var(machine_B, ddof=1)

# Put larger variance in numerator (standard practice)
if var_A >= var_B:
    F_stat = var_A / var_B
    df1 = len(machine_A) - 1
    df2 = len(machine_B) - 1
else:
    F_stat = var_B / var_A
    df1 = len(machine_B) - 1
    df2 = len(machine_A) - 1

# Two-tailed p-value using F distribution
p_value = 2 * min(
    f.cdf(F_stat, df1, df2),
    1 - f.cdf(F_stat, df1, df2)
)

print("F-statistic:", F_stat)
print("Degrees of freedom:", (df1, df2))
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject H₀")
    print("Conclusion: Variances are significantly different.")
else:
    print("Fail to Reject H₀")
    print("Conclusion: Variances are equal.")
```

One-Way ANOVA

📌 Problem

Compare the mean exam scores of students taught using three different methods.

Data
method_A = [78, 82, 85, 80, 79]

method_B = [88, 90, 92, 89, 91]

method_C = [75, 77, 76, 74, 78]




Hypotheses

H₀: μA = μB = μC

H₁: At least one mean is different

```python

method_A = [78, 82, 85, 80, 79]

method_B = [88, 90, 92, 89, 91]

method_C = [75, 77, 76, 74, 78]
```

```python
from scipy.stats import f_oneway

F_stat, p_value = f_oneway(method_A, method_B, method_C)

print("One-Way ANOVA")
print("F-statistic:", F_stat)
print("P-value:", p_value)
```

**Example 1: A/B Test for Conversion Rate (Z-Test)**

| Group | Visitors | Conversions |
| ----- | -------- | ----------- |
| A     | 1000     | 120         |
| B     | 1000     | 150         |


Example: Website Conversion Rate

H₀ (Null): Conversion rate of A = Conversion rate of B

H₁ (Alternative): Conversion rate of A ≠ Conversion rate of B

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Data
conversions = np.array([120, 150])
visitors = np.array([1000, 1000])

# Z-test
z_stat, p_value = proportions_ztest(conversions, visitors)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

Interpretation

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0 → Significant difference")
else:
    print("Fail to reject H0 → No significant difference")
```

Since the p-value is less than 0.05, we reject the null hypothesis and conclude that there is a statistically significant difference between versions A and B. Additionally, because Version B has a higher [conversion rate / mean revenue], we conclude that Version B performs better.”
