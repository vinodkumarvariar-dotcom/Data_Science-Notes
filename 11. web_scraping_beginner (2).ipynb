# Web Scraping Basics – Beginner Notebook
This notebook demonstrates basic web scraping operations using a beginner‑friendly website.

```
from bs4 import BeautifulSoup
import requests

url = "https://quotes.toscrape.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
soup
```

## Extract `<h1>` Tag

```
h1_tag = soup.find('h1')
h1_tag, h1_tag.get_text()
```

## Extract `<p>` Tags

```
p_tags = soup.find_all('p')
p_tags
```

## Extract Text From Paragraphs

```
[p.get_text().strip() for p in p_tags[:5]]
```

## Extract `<a>` Tags

```
a_tags = soup.find_all('a')
a_tags[:10]
```

## Extract Link Text & Href

```
[(tag.get_text(), tag.get('href')) for tag in a_tags[:10]]
```

## Extract Quotes Using Class Selector

```
quotes = soup.find_all('span', class_='text')
[quote.get_text() for quote in quotes[:5]]
```

## Extract Authors

```
authors = soup.find_all('small', class_='author')
[auth.get_text() for auth in authors[:5]]
```

## Using CSS Selectors

```
quote_texts = soup.select('.quote span.text')
[q.get_text() for q in quote_texts[:5]]
```

## Extract Tags for Each Quote

```
tags = soup.select('.quote .tags .tag')
[tag.get_text() for tag in tags[:10]]
```

## Demonstrate `get()` for Link Attributes

```
first_link = soup.find('a')
first_link.get('href')
```
