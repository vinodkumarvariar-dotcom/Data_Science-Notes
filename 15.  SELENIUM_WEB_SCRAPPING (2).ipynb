```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

url = "https://www.amazon.in/s?k=mobiles+under+10000"
driver.get(url)
time.sleep(3)

names, prices, ratings, links = [], [], [], []

for page in range(1, 3):
    driver.get(f"https://www.amazon.in/s?k=mobiles+under+10000&page={page}")
    time.sleep(3)
    products = driver.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")

    for p in products:
        try:
            name = p.find_element(By.TAG_NAME, "h2").text
        except:
            name = None
        try:
            price = p.find_element(By.CLASS_NAME, "a-price-whole").text
        except:
            price = None
        try:
            rating = p.find_element(By.XPATH, ".//span[@class='a-icon-alt']").text
        except:
            rating = None
        try:
            link = p.find_element(By.TAG_NAME, "a").get_attribute("href")
        except:
            link = None

        names.append(name)
        prices.append(price)
        ratings.append(rating)
        links.append(link)

driver.quit()

df = pd.DataFrame({
    "Product Name": names,
    "Price (₹)": prices,
    "Rating": ratings,
    "Link": links
})

df.to_csv("amazon_mobiles_under_10000.csv", index=False)
print("✅ Data saved successfully to 'amazon_mobiles_under_10000.csv'")
```
