```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Set plot style for better aesthetics
plt.style.use('ggplot')

# --- Code ---
CSV_FILE = 'marketing_data.csv'
df = pd.read_csv(CSV_FILE)
df
```

2. 📈 Line Chart (Trend Analysis - Sales over Time)
This chart shows the continuous change between two variables.

```python
# CODE CELL 2: Line Chart
# --- Code ---
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['Sales_Region_A'], marker='o', linestyle='-', color='teal', linewidth=2, label='Sales')

plt.xlabel('Month')
plt.ylabel('Sales Volume ($ thousands)')
plt.title('Line Chart: Sales Trend for Region A')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Explanation of Code Cell 2
This cell generates a Line Chart, best for visualizing trends over an ordered sequence (time).

plt.figure(figsize=(12, 6)): Creates a new figure object with a size of 12 inches wide by 6 inches tall.

plt.plot(df['Month'], df['Sales_Region_A'], ...): The core function, plotting the Sales_Region_A values against the Month labels.

marker='o' displays a circle at each data point.

linestyle='-' ensures the points are connected by a solid line.

plt.xlabel(), plt.ylabel(), plt.title(): Add context to the plot.

plt.grid(True, ...): Adds grid lines for easier reading of the values.

plt.show(): Displays the finished plot.

3. 📉 Scatter Plot (Relationship - Visits vs. Sales)
This chart helps visualize the relationship between two numerical variables.

```python
#Scatter Plot

plt.figure(figsize=(10, 6))
# Using Satisfaction for color (c) and Expenses for size (s)
plt.scatter(df['Visits'], df['Sales_Region_A'], s=df['Expenses']*5, c=df['Satisfaction'], cmap='viridis', alpha=0.8)

plt.colorbar(label='Average Customer Satisfaction Score')
plt.xlabel('Website Visits')
plt.ylabel('Sales Region A')
plt.title('Scatter Plot: Relationship between Visits and Sales (Multi-variate)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

Explanation of Code Cell 3
This cell creates a Scatter Plot, used to investigate the correlation or relationship between two continuous variables.

plt.scatter(df['Visits'], df['Sales_Region_A'], ...): Plots individual points with Visits on the X-axis and Sales_Region_A on the Y-axis.

s=df['Expenses']*5: Maps the point size (s) to the Expenses column, allowing the visualization of a third variable.

c=df['Satisfaction']: Maps the point color (c) to the Satisfaction column, introducing a fourth variable.

cmap='viridis': Specifies the colormap to use for the color variable.

plt.colorbar(...): Provides a key for the color mapping (Satisfaction Score).

4. 📊 Bar Chart (Comparison - Expenses per Month)
This chart compares quantities across discrete categories.

```python
# CODE CELL 4: Bar Chart
# --- Code ---
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Expenses'], color='darkorange', edgecolor='black')

plt.xlabel('Month')
plt.ylabel('Expenses ($ thousands)')
plt.title('Bar Chart: Monthly Expenses')
plt.xticks(rotation=0) # Keeps month names horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

This cell generates a Bar Chart, excellent for comparing discrete categories.

plt.bar(df['Month'], df['Expenses'], ...): Plots bars where the height is determined by Expenses and the label is set by Month.

edgecolor='black': Adds a border to each bar for visual separation.

plt.xticks(rotation=0): Sets the rotation of the X-axis labels to prevent overlap (though not strictly necessary for 3-letter months).

plt.grid(axis='y', ...): Only shows grid lines along the Y-axis to help read the bar heights without cluttering the chart.

5. 🥧 Pie Chart (Proportion - Visits by Campaign Group)
This chart shows how different parts contribute to a total.

```python
# CODE CELL 5: Pie Chart
# --- Code ---
# Aggregate data: sum up Visits for each Campaign Group
pie_data = df.groupby('Campaign_Group')['Visits'].sum()
labels = pie_data.index
sizes = pie_data.values
colors = ['lightcoral', 'lightskyblue']
explode = (0.05, 0) # Slightly separate the first slice

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.title('Pie Chart: Total Visits by Campaign Group')
plt.axis('equal') # Ensures the pie is drawn as a perfect circle
plt.show()
```

pie_data = df.groupby(...)['Visits'].sum(): Groups the DataFrame by Campaign_Group and calculates the total Visits for each group, which is necessary before plotting the proportion.

plt.pie(sizes, ...): The core function plots the pie chart.

autopct='%1.1f%%': Formats the percentage displayed on each slice to one decimal place.

explode: An array used to offset (or "explode") a slice, drawing attention to it.

plt.axis('equal'): Required to ensure the pie chart is circular, not oval.

6. 🔊 Histogram (Distribution - Satisfaction Scores)
This chart shows the frequency distribution of a continuous variable.

```python
# CODE CELL 6: Histogram
# --- Code ---
plt.figure(figsize=(10, 6))
plt.hist(df['Satisfaction'], bins=5, color='darkseagreen', edgecolor='black', alpha=0.9)

plt.xlabel('Satisfaction Score (Binned)')
plt.ylabel('Frequency (Count of Months)')
plt.title('Histogram of Customer Satisfaction Score Distribution')
plt.grid(axis='y', linestyle='-', alpha=0.5)
plt.show()
```

This cell generates a Histogram, which shows the frequency distribution of a single, continuous variable (Satisfaction).

plt.hist(df['Satisfaction'], ...): Creates the histogram.

bins=5: Divides the range of Satisfaction scores into 5 intervals (bins) and counts how many data points fall into each bin.

edgecolor='black' and alpha=0.9 enhance the visual clarity of the bars.

The X-axis represents the score ranges (bins), and the Y-axis represents the frequency (how many months had scores in that range).

7. 📦 Box Plot (Summary - Comparison of Sales)
This chart summarizes the five-number statistical summary of data.

```python
# CODE CELL 7: Box Plot
# --- Code ---
# Prepare data for comparing Sales and Expenses distributions
plot_data = [df['Sales_Region_A'], df['Expenses']]
labels = ['Sales', 'Expenses']

plt.figure(figsize=(8, 6))
plt.boxplot(plot_data, vert=True, patch_artist=True,
            labels=labels,
            boxprops={'facecolor': 'lightblue'},
            medianprops={'color': 'red', 'linewidth': 2})

plt.ylabel('Value ($ thousands)')
plt.title('Box Plot: Comparison of Sales vs. Expenses Distribution')
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.show()
```

This cell creates a Box Plot (or Box-and-Whisker Plot), used to quickly visualize the statistical summary and distribution of data.

plot_data = [df['Sales_Region_A'], df['Expenses']]: Prepares a list of the two columns we want to compare side-by-side.

plt.boxplot(plot_data, ...): Generates the box plot.

vert=True: Creates vertical boxes (default).

patch_artist=True: Allows for custom coloring of the box interior.

medianprops: Customizes the line representing the median (the 50th percentile).

The plot shows the minimum, Q1 (25th percentile), Median, Q3 (75th percentile), and maximum (whiskers) for both Sales and Expenses.
